```python
import matplotlib.pyplot as plt
from matplotlib import cm
import pandas as pd
#import cufflinks as cf
import seaborn as sns
import numpy as np
import random
sns.set_style('white')

%matplotlib inline
```

```python
# Load raw data into DataFrame using pandas
vehicle_type_df_raw = pd.read_csv('Type_of_Vehicles_and_Type_of_Licence.csv', delimiter=',')
vehicle_type_df_raw[:2]
```

```python
#replace NaN with 0
vehicle_type_df = vehicle_type_df_raw[~(vehicle_type_df_raw['SNo'] == 'Total')].fillna(0)
vehicle_type_total_df = vehicle_type_df_raw[vehicle_type_df_raw['SNo'] == 'Total']

#validation
vehicle_type_df.head()
```

```python
# Final
#ToDo : tooltip issue
#Stacked bar plot with bokeh

from bokeh.plotting import figure, show, output_file, ColumnDataSource
from bokeh.palettes import Spectral6
from bokeh.core.properties import value
from math import pi
from bokeh.models import HoverTool

sorted_license_df = vehicle_type_df.sort_values(['RegularLicence-Accident-2014'], ascending=[False])
sorted_license_top20_df = sorted_license_df[:20]

states = list(sorted_license_top20_df['States'])
regular = list(sorted_license_top20_df['RegularLicence-Accident-2014'])
learner = list(sorted_license_top20_df['LearnersLicence-Accident-2014'])
without = list(sorted_license_top20_df['WithoutLicence-Accident-2014'])
print(states)

license_type = ['Regular', 'Learner', 'NoLicense']
colors = ['#2b83ba', '#abdda4', '#ffffbf']

data = {'States' : states,
        'Regular'   : regular,
        'Learner'   : learner,
        'NoLicense'   : without}

source = ColumnDataSource(data=data)

hover = HoverTool(tooltips=[
    ('Regular', '@Regular'),
    ('Learner', '@Learner'),
    ('No License', '@NoLicense')
])


p = figure(x_range=states, plot_height=600, plot_width=800, title="License Types of Driver vs Accidents",
           toolbar_location=None, tools=[hover],
           x_axis_label="States", y_axis_label="No of Accidents")

p.vbar_stack(license_type, x='States', width=0.9, color=colors, source=source,
             legend=[value(x) for x in license_type])

p.xaxis.major_label_orientation = "vertical"
p.xaxis.axis_line_color = "blue"
p.xaxis.axis_label_text_color = "blue"
p.yaxis.axis_label_text_color = "blue"
p.title.align = "center"
p.legend.location = "top_right"
p.legend.orientation = "horizontal"
p.y_range.start = 0.1
p.x_range.range_padding = 0.1
p.axis.minor_tick_in = -3
p.axis.minor_tick_out = 6

show(p)
```
