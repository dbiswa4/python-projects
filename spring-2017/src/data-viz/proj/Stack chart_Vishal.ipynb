```python
# import modules
import pandas as pd

# Import the excel file and call it xls_file
xls_file = pd.ExcelFile('Updated_Data.xls')
xls_file
# View the excel file's sheet names
xls_file.sheet_names

df = xls_file.parse('Worksheet')
df
#df.Education.unique()
```

```python
%pylab inline
import pandas as pd
import matplotlib.pyplot as plt
test5 = df.groupby(['States', 'Education'])['States'].count().unstack('Education').fillna(0)

test5.plot(kind='bar', stacked=True)
```

```python
graph_by_duration = df.pivot(index='States', columns='Education', values='Total')
ax = graph_by_duration.plot.bar(figsize=(17,8), width=2)
ax.set_ylabel('Accidents')
ax.set_title('Accidents per State')
plt.show()
```

```python
from bokeh.charts import Bar, show, output_notebook, output_file

output_notebook()
Education = df.Education.tolist()
States = df.States.tolist()
Total = df.Total.tolist()


data_dict = { 'Accidents' : Total,
              'States' : States,
              'Education' : Education
            }

data_df = pd.DataFrame(data_dict)

p = Bar(data_df, 
        values='Accidents', 
        label='States', 
        stack='Education',
        legend='top_right',
        title="      Accident Count in India by State and Education",
        tooltips=[ 
                  ('States', '@States'), 
                  ('Education', '@Education'),
        ('Accidents', '@height out of Accidents')]
       )

#output_file("Programming_Skills_ECEn360_Win2016.html")

show(p)
```

```python
# a stacked bar plot with errorbars
import numpy as np
import matplotlib.pyplot as plt


N = 5
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
menStd = (2, 3, 4, 1, 2)
womenStd = (3, 5, 2, 3, 3)
ind = np.arange(N)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence

p1 = plt.bar(ind, menMeans, width, color='#d62728', yerr=menStd)
p2 = plt.bar(ind, womenMeans, width,
             bottom=menMeans, yerr=womenStd)

plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0]), ('Men', 'Women'))

plt.show()
```
