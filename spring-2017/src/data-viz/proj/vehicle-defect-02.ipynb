```python
# Load raw data into DataFrame using pandas
vehicle_defect_df_raw = pd.read_csv('vehicle_defect.csv', delimiter=',')
```

```python
#replace NaN with 0
vehicle_defect_df = vehicle_defect_df_raw[~(vehicle_defect_df_raw['State'] == 'Total')].fillna(0)
vehicle_defect_total_df = vehicle_defect_df_raw[vehicle_defect_df_raw['State'] == 'Total']
```

```python
# Final
#Stacked bar plot with bokeh

from bokeh.plotting import figure, show, output_file, ColumnDataSource
from bokeh.palettes import Spectral6
from bokeh.core.properties import value
from math import pi
from bokeh.models import HoverTool

output_file("vehicle-defect-accidents.html")

sorted_total_accidents_df = vehicle_defect_df.sort_values(['DefectiveBrakes-NumberofAccident-2014'], ascending=[False])
sorted_total_accidents_top20_df = sorted_total_accidents_df[:20]

sorted_total_accidents_df[:2]

states = list(sorted_total_accidents_top20_df['State'])
brakes = list(sorted_total_accidents_top20_df['DefectiveBrakes-NumberofAccident-2014'])
steering = list(sorted_total_accidents_top20_df['DefectiveSteering-NumberofAccident-2014'])
bursttyres = list(sorted_total_accidents_top20_df['PuncturedBurstTypres-NumberofAccident-2014'])
baldtyres = list(sorted_total_accidents_top20_df['BaldTyres-NumberofAccident-2014'])
othermechanical = list(sorted_total_accidents_top20_df['OtherSeriousMechanicalDefect-NumberofAccident-2014'])
print(states)

defect_type = ['Brakes', 'Steering', 'BurstTyres', 'BaldTyres', 'OtherMechanical']
colors = ['#5e4fa2', '#bebada', '#fb8072', '#80b1d3', '#bc80bd']
#colors = ['#b3e2cd', '#fdcdac', '#cbd5e8', '#f4cae4', '#e6f5c9']
#colors = ['#993404', '#d95f0e', '#fe9929', '#fed98e', '#ffffd4']

data = {'States' : states,
        'Brakes'   : brakes,
        'Steering'   : steering,
        'BurstTyres'   : bursttyres,
        'BaldTyres': baldtyres,
       'OtherMechanical' : othermechanical}

source = ColumnDataSource(data=data)

hover = HoverTool(tooltips=[
    ('Brakes', '@Brakes'),
    ('Steering', '@Steering'),
    ('Burst Tyres', '@BurstTyres'),
    ('Bald Tyres', '@BaldTyres'),
    ('Other Mechanical', '@OtherMechanical')
])


p = figure(x_range=states, plot_height=600, plot_width=800, title="Accidents Cause",
           toolbar_location=None, tools=[hover],
           x_axis_label="States", y_axis_label="No of Accidents")

p.vbar_stack(defect_type, x='States', width=0.9, color=colors, source=source,
             legend=[value(x) for x in defect_type])

p.xaxis.major_label_orientation = "vertical"
p.xaxis.axis_line_color = "blue"
p.xaxis.axis_label_text_color = "orange"
p.yaxis.axis_label_text_color = "orange"
p.title.align = "center"
p.legend.location = "top_right"
p.legend.orientation = "horizontal"
p.y_range.start = 0.1
p.x_range.range_padding = 0.1
p.axis.minor_tick_in = -3
p.axis.minor_tick_out = 6

show(p)
```
