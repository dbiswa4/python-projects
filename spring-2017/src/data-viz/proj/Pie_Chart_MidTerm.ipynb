```python
import plotly.plotly as py
import plotly.graph_objs as go

xls_file = pd.ExcelFile('Updated_Data_Pie.xls')
xls_file
# View the excel file's sheet names
xls_file.sheet_names

df = xls_file.parse('Worksheet')
fig = {
  "data": [
    {
      "values": df.Total ,
      "labels":df.Education,
      'marker': {'colors': ['rgb(146, 123, 21)',
                                  'rgb(177, 180, 34)',
                                  'rgb(206, 206, 40)',
                                  'rgb(175, 51, 21)',
                                  'rgb(35, 36, 21)']},
      "domain": {"x": [.20, 1]},
      "name": "Accident",
      "hoverinfo":"label+percent+name",
      "hole": .4,
      "type": "pie"
    },     
    ],
  "layout": {
        "title":"Accident Count By Education Type",
        "annotations": [
            {
                "font": {
                    "size": 20
                },
                "showarrow": False,
                "text": "States",
                "x": 0.60,
                "y": 0.5
            },
            
        ]
    }
}
py.iplot(fig, filename='pie_chart_subplots')
```
