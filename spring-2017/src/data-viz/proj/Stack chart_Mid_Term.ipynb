```python
# import modules
import pandas as pd

# Import the excel file and call it xls_file
xls_file = pd.ExcelFile('Updated_Data.xls')
xls_file
# View the excel file's sheet names
xls_file.sheet_names
ds = xls_file.parse('Worksheet')
df = ds[(ds['States'] == 'Uttar Pradesh') | (ds['States'] ==  'Telangana')
      | (ds['States'] ==  'Andhra Pradesh')| (ds['States'] ==  'Rajasthan') |  (ds['States'] ==  'Tamil nadu')  ]
df
#df.Education.unique()
```

```python

from bokeh.charts import Bar, show, output_notebook, output_file


output_notebook()
Education = df.Education.tolist() 
States = df.States.tolist()
Total = df.Total.tolist()




data_dict = { 'Accidents' : Total,
              'States' : States,
              'Education' : Education
            }

data_df = pd.DataFrame(data_dict)

p = Bar(data_df, 
        values='Accidents', 
        label='States', 
        stack='Education',
        legend='top_right',
        title="      Accident Count in India by State and Education",
        tooltips=[ 
                  ('States', '@States'), 
                  ('Education', '@Education'),
        ('Accidents', '@height out of Accidents')]
       )

#output_file("Programming_Skills_ECEn360_Win2016.html")

show(p)
```
